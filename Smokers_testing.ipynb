```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# load the dataset
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
df = pd.read_csv(url)
```

```python
# Analyze the relationship between smoking and charges
smoking_charges = df[['smoker', 'charges']]
```

```python
# Group data by smoking status and calculate average charges
avg_charges = smoking_charges.groupby('smoker').mean()
print("Average charges by smoking status:\n", avg_charges)
```

```python
# Visualize the relationship between smoking and medical costs
plt.figure(figsize=(8, 6))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title('Boxplot of Charges by Smoking Status')
plt.show()
```

```python
# Perform statistical test  if the difference in charges between smokers and non smokers is significant
smokers = df[df['smoker'] == 'yes']['charges']
non_smokers = df[df['smoker'] == 'no']['charges']
```

```python
# Perform t-test
t_stat, p_val = stats.ttest_ind(smokers, non_smokers)
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```
